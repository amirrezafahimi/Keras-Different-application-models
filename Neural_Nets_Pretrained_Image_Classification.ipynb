```
import numpy as np
import tensorflow as tf
import keras 

from keras.preprocessing.image import img_to_array, load_img
from keras.applications.inception_resnet_v2 import InceptionResNetV2
from keras.applications.inception_resnet_v2 import decode_predictions as inception_decoder
from keras.applications.inception_resnet_v2 import preprocess_input as preprocess_inception

from keras.applications.vgg19 import VGG19
from keras.applications.vgg19 import decode_predictions as vgg_decoder
from keras.applications.vgg19 import preprocess_input as preprocess_vgg

from keras.applications.nasnet import NASNetLarge
from keras.applications.nasnet import preprocess_input as nasnet_preprocess
from keras.applications.nasnet import decode_predictions as nasnet_decoder

from IPython.core.display import display
```

```
model_inception = InceptionResNetV2(weights='imagenet')
model_vgg = VGG19(weights='imagenet')
nasnet_model = NASNetLarge(weights='imagenet')
```

```
def inception_resnet_method(file):
  print('Here is the InceptionResNet model prediction on the given image \n')
  raw_img = load_img(file, target_size=(299, 299))
  display(raw_img)
  array_pic = img_to_array(raw_img)

  preprocessed_img = preprocess_inception(array_pic)
  expand_img = np.expand_dims(preprocessed_img, axis=0)
  preds = model_inception.predict(expand_img)
  return inception_decoder(preds)  
```

```
def vgg_model(file):
  print('Here is the VGG19 model prediction on the given image \n')
  raw_img = load_img(file, target_size=(224, 224))
  display(raw_img)
  array_pic = img_to_array(raw_img)

  preprocessed_img = preprocess_vgg(array_pic)
  expand_img = np.expand_dims(preprocessed_img, axis=0)
  preds = vgg19_model.predict(expand_img)
  return vgg_decoder(preds)
```

```
def nasnet_model_prediction(file):
  print('Here is the NASNetLarge model prediction on the given image \n')
  raw_img = load_img(file, target_size=(331, 331))
  display(raw_img)
  array_pic = img_to_array(raw_img)

  preprocessed_img = nasnet_preprocess(array_pic)
  expand_img = np.expand_dims(preprocessed_img, axis=0)
  preds = nasnet_model.predict(expand_img)
  return nasnet_decoder(preds)
```
